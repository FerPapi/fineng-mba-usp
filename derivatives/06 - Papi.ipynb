## Aluno: Fernando Gomes Papi

A diferença entre os preços de contratos futuros e contratos a termo ocorre devido à correlação entre o preço do ativo subjacente e as taxas de juros e a preferência temporal do investidor. 

Quando há uma correlação positiva, os contratos futuros são ligeiramente mais caros do que os contratos a termo, devido aos ganhos e perdas imediatas associados à variação das taxas de juros. 

Se um trader tem uma posição comprada no contrato futuro, e S é positivamente correlacionado com a taxa de juros, um aumento em S significa um ganho imediato ao trader devido ao settlement diário. 

O ganho será então reinvestido a uma taxa de juros mais alta, e a situação oposta também ocorre. 

Se S cai, a perda será refinanciada a uma taxa de juros mais baixa. Por prover uma utilidade ligeiramente maior, os preços do futuro tendem a ser um pouco maiores do que do forward.

Quando a correlação é negativa, a situação se inverte, e os contratos a termo são ligeiramente mais caros do que os contratos futuros. 

Para contratos curtos, de apenas alguns meses, estas diferenças provavelmente serão pequenas a ponto de poderem ser ignoradas. Porém para contratos mais longos, pode-se perceber a diferença.

Além disso, em geral, contratos futuros têm mais liquidez e menor risco de contra-parte
