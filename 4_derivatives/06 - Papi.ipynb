## Aluno: Fernando Gomes Papi

A diferena entre os preos de contratos futuros e contratos a termo ocorre devido  correlao entre o preo do ativo subjacente e as taxas de juros e a preferncia temporal do investidor. 

Quando h uma correlao positiva, os contratos futuros so ligeiramente mais caros do que os contratos a termo, devido aos ganhos e perdas imediatas associados  variao das taxas de juros. 

Se um trader tem uma posio comprada no contrato futuro, e S  positivamente correlacionado com a taxa de juros, um aumento em S significa um ganho imediato ao trader devido ao settlement dirio. 

O ganho ser ento reinvestido a uma taxa de juros mais alta, e a situao oposta tambm ocorre. 

Se S cai, a perda ser refinanciada a uma taxa de juros mais baixa. Por prover uma utilidade ligeiramente maior, os preos do futuro tendem a ser um pouco maiores do que do forward.

Quando a correlao  negativa, a situao se inverte, e os contratos a termo so ligeiramente mais caros do que os contratos futuros. 

Para contratos curtos, de apenas alguns meses, estas diferenas provavelmente sero pequenas a ponto de poderem ser ignoradas. Porm para contratos mais longos, pode-se perceber a diferena.

Alm disso, em geral, contratos futuros tm mais liquidez e menor risco de contra-parte
