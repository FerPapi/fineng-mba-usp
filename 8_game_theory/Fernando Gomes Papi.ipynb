# Um modelo duopolista de publicidade de plataformas de apostas esportivas online

## Introdução 

Recentemente, temos visto um grande crescimento do mercado de apostas esportivas on line (OBS - online bet systems). Estas OBS competem ferozmente por espaços publicitários ligados ao esporte, especialmente no Futebol, onde particularmente nota-se a sua grande presença em placas, anúncios de TV, uniformes dos times, e mais. Das 20 equipes participantes do campeonato brasileiro, 19 possuem patrocínios masters ligados a estas empresas de apostas, e ainda por cima, os contratos estão quebrando recordes históricos a cada temporada. [[1]](https://www.lance.com.br/lancebiz/sites-de-apostas-dominam-mercado-de-patrocinio-master-no-futebol-brasileiro-veja-raio-x.html)

A motivo da busca desenfreada por atenção dos apostadores pode ser hipoteticamente descrito por alguns fatores. 

O primeiro e mais importante é o reconhecimento da marca. Muitas plataformas online de apostas são fraudulentas, e com inúmeras competidoras, a marca precisa garantir que o público a reconheça como confiável, atrelando sua marca a um clube de futebol ou competição esportiva. Quanto mais tempo no mercado publicitário, mais a marca será vista como confiável. 

Na mesma linha, os produtos oferecidos pelas plataformas são muito semelhantes e com pouca margem de diferenciação. Assim, o consumidor tenderá a escolher, dentre as confiáveis, uma marca que está estabelecida e é facilmente reconhecida.

Além disso, o investimento em publicidade faz crescer o próprio mercado de apostas como um todo, garantido um mercado maior a cada ano.

Este é um jogo de decisão sequencial com informação completa, e usaremos o Duopólio de Cournot-Stackelberg para propor uma solução para o problema: *quanto alocar em anúncios esportivos em um jogo de competição duopolista?*

## Formulação do problema:

Seja o mercado de apostas esportivas um mercado duopolista, com firmas $i=1$ e $j=2$. 

A partir do modelo de duopólio de Cournot-Stackelberg, propõe-se modificações para o problema. 

Não há um custo de bens produzidos, já que o custo da operação de plataformas online é majoritariamente fixo. Além disso, nenhuma das empresas consegue impor um custo ao mercado, pois as apostas podem ser consideradas como gratuitas. O "custo" do jogador está atrelado ao seu desempenho como apostador e ao payoff (às _odds_) das plataformas, e portanto, não há influência de custo no mercado total de apostas. Poderemos considerar a demanda total $D$ como fixa para um período, e, a rigor, a inclusão ou exclusão de um player do mercado não afetará a demanda total. A competição das empresas é, portanto, para capturar a maior fatia do mercado possível dentro da relação entre a demanda e o total capturável por anúncios. 

A firma $i$ decide o quanto gastará em anúncios, $a_1$. A firma $j$ fica sabendo do valor investido pretendido e decide investir $a_2$. Sendo um jogo de informação completa, $i$ sabe desta informação para investir $a_1$ baseado em $a_2$.

Ambas tem conhecimento da demanda total $D$ para o período, ou seja, quanto esperam que a indústria como um todo arrecade com apostas.

O investimento total em anúncios das firmas é $A = a_1 + a_2$. 

Podemos assumir que há uma relação direta e linear da quantidade investida e do percentual de $D$ que cada empresa conseguirá capturar.

Assim, podemos modelar a função de lucro das firmas como:
$$
\Pi_i = \frac{a_i}{a_i+a_j} * D * p - a_i
$$

Ou seja, a proporção do mercado de apostas capturado pela firma é diretamente proporcional à sua participação no mercado publicitário. 

A constante $p = [0,1)$ é a proporção que a empresa retém como ganhos de cada dólar que entra em seu sistema, e a empresa consegue controlar este valor através da sua gestão de risco, definição das probabilidades dos eventos e seus pay-offs para o apostador. Podemos assumir que em um mercado equilibrado e competitivo, a constante $p$ é a mesma para todos os jogadores no longo prazo, pois nenhum consegue ofertar _odds_ muito discrepantes do restante do mercado por questões de gerência de risco. _Odds_ muito atrativas atrairiam mais jogadores, o que *aumenta* o risco de payoffs excessivos em casos de eventos adversos, mal-precificados ou que tiveram um risco mal calculado.

De maneira resumida, o lucro da empresa é o quanto ela conseguir capturar do mercado total (através de publicidade) vezes um fator de lucro interno menos seus gastos com publicidade. Não levaremos em consideração custos fixos e eventuais receitas fixas, como taxas.

Podemos definir $r = D*p$, e portanto, podemos escrever o lucro também como:
$$
\Pi_i = \frac{-a_i^2 + a_i*(r-a_j)}{a_i+a_j}
$$

A partir da formulação do lucro da firma, podemos obter sua alocação ótima $a_i^*$

## Solução do problema:

A solução do problema se dá encontrando um equilírbio de Nash perfeito em sub-jogos. Podemos fazê-lo por backward induction.

### Resolvendo $a_2^*$

A firma $j=2$ terá sua melhor resposta à firma $i=1$ quando:

$$
\Pi_2(a_1, a_2^*) = \max{\frac{-a_2^2 + a_2*(r-a_1)}{a_1+a_2}}
$$

A derivada de primeira ordem estabelece que:
$$
\frac{\delta{\Pi_2}}{\delta{a_2}} = 0
$$

$$
\frac{-a_2^2 + 2a_2a_1 + (a_1^2 - ra_1)}{(a_1+a_2)^2} = 0
$$

$$
\implies a_2^* = -a_1 + \sqrt{ra_1}
$$

é a quantidade ótima que a firma 2 deve investir em anúncios

### Resolvendo $a_1^*$

A firma 1 deverá antecipar que a firma 2 jogará de acordo com sua decisão (escolhendo a primeira equação do lucro pela facilidade de manipulação algébrica). Assim,

$$
\begin{aligned}
\Pi_1(a_1^*, a_2^*) &= \max{\frac{a_1}{a_1+ -a_1 + \sqrt{ra_1}} * D * p - a_1}\\
&= \max{\frac{ra_1}{\sqrt{ra_1}} - a_1}
\end{aligned}
$$

A derivada de primeira ordem estabelece que:
$$
\frac{\delta{\Pi_1}}{\delta{a_1}} = 0
$$

$$
\frac{r}{2\sqrt{ra_1}} - 1 = 0
$$

$$
\implies a_1^* = \frac{r}{4}
$$


### Simplificando resultados:

Temos que 

$$
a_1^* = \frac{r}{4}
$$

e substituido em $a_2$, é fácil verificar que:

$$
a_2^* = -\frac{r}{4} + \sqrt{r*\frac{r}{4}} = \frac{r}{4}
$$

e 

$$
a_1^* = a_2^*
$$

Ou seja, as duas firmas dividirão perfeitamente o duopólio de anúncios e o mercado, sem incentivos para concorrência acirrada. 

#### Resultados e Simulações numéricas

Podemos entender melhor o jogo através de simulações numéricas em torno do ponto ótimo, tornando-o um jogo repetido de informação completa.

Consideremos um mercado grande, onde $D = 100 M$, i.e, 100 milhões de dólares. E a taxa de lucro do mercado é $p=0.5$

Então a quantidade ótima para ambos os jogadores será de $a = \frac{100*10^6*0.5}{4} = 250.000$ dólares. 

Mesmo começando com quantidades muito inferiores, ao tornar este jogo repetido (mantendo-se todas as condições), rapidamente os jogadores convergem para o valor ótimo.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
import numpy as np
from numpy import power

def ai_opt(aj, D, p):
    # best response for second mover
    return (-aj + np.sqrt(D*p*aj))

def a1_opt(D, p):
    # best response for first mover
    return D*p/4

def profit_i(ai, aj, D, p):
    # profit for company y 
    return int(ai/(ai+aj) * D * p - ai)

def opt_profit_a1(a, D, p):
    # optimal profit for first mover
    r = D*p
    profit = (-a + np.sqrt(r*a))*r/(np.sqrt(r*a)) - a
    return profit
```

```python
D = 100e6
p = 0.50

a1_s = a1_opt(D, p)
a2_s = ai_opt(a1_s, D, p)

profit_a1 = profit_i(a1_s, a2_s, D, p)
profit_a2 = profit_i(a2_s, a1_s, D, p)

print(f'{profit_a1 = }', f'{a1_s = }')
print(f'{profit_a2 = }', f'{a2_s = }')

# simulando um jogo repetido
p1 = []
p2 = []
a1 = 10
a2 = 15

for i in np.linspace(0, 1, 5):    
    payoff_1 = profit_i(a1, a2, D, p)
    payoff_2 = profit_i(a2, a1, D, p)

    p1.append(a1)
    p2.append(a2)

    a1 = max(0, ai_opt(a2, D, p))
    a2 = max(0, ai_opt(a1, D, p))
    if i %2 == 0:
        a1 = max(0, ai_opt(a2, D, p))

plt.figure(figsize=(9,4))
plt.plot(p1, label = '1')
plt.plot(p2, label = '2')
plt.axhline(y=opt_profit_a1(a1_s, D, p), color='black', linestyle='--', label='optimal', lw=2)
plt.legend()
plt.grid()
plt.show()
```

Aqui, podemos entender o que acontece se um jogador desviar do ponto ótimo. Ao variar a quantidade do jogador 1, o jogador 2 tenta obter a melhor resposta. Conforme o jogo se aproxima do valor ótimo, o lucro de ambos se iguala. Porém, ao jogar um valor superior ao ótimo, o jogador 1 comça a ter um decaimento lento do seu payoff. Em contrapartida, o jogador 2 passa a ter um domínio muito menor do mercado e vê o seu payoff decair drasticamente.

Esta pode ser uma tática usada pelo jogador 1 para tirar o jogador 2 do mercado. Em contrapartida, se o jogador 2 conseguir entender melhor o comportamento do jogador 1, poderá modelar o jogo de forma a obter uma melhor resposta.

```python
D = 100e6
p = 0.50

p1 = []
p2 = []

for i in range(5_750_000, 17_750_000, 100000):
    a1 = i
    a2 = max(0, ai_opt(a1, D, p))
    payoff_1 = profit_i(a1, a2, D, p)
    payoff_2 = profit_i(a2, a1, D, p)

    p1.append(payoff_1)
    p2.append(payoff_2)

plt.figure(figsize=(9,4))
plt.plot(p1, label = '1')
plt.plot(p2, label = '2')
plt.axhline(y=opt_profit_a1(a1_s, D, p), color='black', linestyle='--', label='optimal', lw=2)
plt.legend()
plt.grid()
plt.show()
```

Neste caso, poderíamos simular o crescimento do mercado de acordo com o investimento feito no período anterior. Isto explicaria o crescente investimento feito por empresas do setor ano após ano. Mais anúncios trazem mais usuários, e para não ficar para trás na competição, as empresas anunciam mais, gerando um efeito de bolha até o mercado saturar

### Dicussão dos resultados e Conclusão

O modelo apresentado, ainda que simplista, evidencia o motivo de estarmos acompanhando um crescimento exponencial do mercado de anúncios de empresas do ramo de apostas esportivas. Num contexto mais simples, em que 2 empresas competem por uma demanda fixa através de anúncios, o equilíbrio de Nash sugere que ambas devem repartir o mercado para maximizar seus lucros. Mesmo que as empresas comecem com investimentos modestos, logo tendem ao equilíbrio de nash para capturar todo o mercado, se extrapolarmos este jogo para um jogo repetido com informação completa.

Evidenciado pela segunda simulação apresentada, haveria incentivo para as empresas anunciarem o mesmo valor, porém, o mínimo possível. Assim ambas teriam 50% do mercado, porém com margem muito maior. Isto, entretanto, não é um equilíbrio de nash, pois uma empresa que quebrasse o cartel e anunciasse mais, receberia uma fatia maior do mercado e obteria um lucro ainda maior. 

No entando, uma empresa que anuncie muito além do valor estabelecido pelo equilíbrio de Nash pode comprometer seriamente os lucros da oura empresa competidora.

Estes efeitos, bem como o crescimento do mercado pelo anúncios após cada período, o impacto de odds mais favoráveis ($p$) ou a presença de mais competidores poderiam ser melhores representados por uma modelagem mais complexa do problema
