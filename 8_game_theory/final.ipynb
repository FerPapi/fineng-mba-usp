Caso o Jogador 3 não atue, então os Jogadores 1 e 2 tem liberdade para atuar, mas só o farão caso o outro jogador não atue. Desta forma, os equilíbrios de Nash serão:

Para uma estratégia mista, se a probabilidades $(q, p, r)$ correspondem à probabilidade de cada jogador atuar, então a estratégia mista pode se resumir ao sistema de equações não lineares:
$$
-3q - 3r + 3qr + 2 = 0
$$

$$
-3p - 3r + 3pr + 2 = 0
$$

$$
-3q - 3p + 3qp + 2 = 0
$$

e resolvendo este sistema via python (abaixo), temos que $q=p=r=0.4226$

```python
from scipy.optimize import fsolve
import math

def equations(a):
    q, r, p = a
    return (-3*q - 3*r + 3*q*r + 2, 
            -3*p - 3*r + 3*p*r + 2,   
            -3*q - 3*p + 3*q*p + 2,
           )

q, r, p =  fsolve(equations, (0.4, 0.4, 0.4))

print(f"{q = :.4f}, {r = :.4f}, {p = :.4f}")
```

Mantendo a modelagem do Duopólio de Cournot para um caso de monopólio, temos que a produção total $Q = q_1$, ou seja, só uma firma produzirá, com custo $c_1 > 0$

O preço é dado por:
$$
p(q_1) = A - q_1
$$

O lucro desta firma será então

$$
\pi_1(q_1) = (A - q_1) * q_1 - c_1 * q_1
$$

o ponto ótimo do lucro será

$$
\frac{\delta\pi}{\delta q} = A - 2q_1^* - c_1 = 0
\implies q_1^* = \frac{A - c_1}{2}
$$

que é um ponto de máximo pois a derivada segunda é negativa.

O payoff, inserindo $q^*$ em $\pi_1(q_1)$ será:

$$
\pi_1(q^*) = Aq^* - {q^*}^2 - c_1q^* = {-q^*}^2 + q^*(A-c_1)
$$

como $(A-c_1) = 2q^*$, teremos:
$$
\pi_1(q^*) = -{q^*}^2 + q^*(2q^*) = {(q^*)}^2
$$

O preço é dado por:
$$
p(Q) = A - Q = A - q_1 - q_2

$$
\pi_i(q_i) = (A - q_i - q_j - c_i) * q_i = -(q_i)^2 + q_i(A - q_j - c_i)

$$

$$
