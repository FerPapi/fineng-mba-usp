Para uma estratégia mista, se a probabilidades $(q, p, r)$ correspondem à probabilidade de cada jogador atuar, então a estratégia mista pode se resumir ao sistema de equações não lineares:
$$
-3q - 3r + 3qr + 2 = 0
$$

$$
-3p - 3r + 3pr + 2 = 0
$$

$$
-3q - 3p + 3qp + 2 = 0
$$

e resolvendo este sistema via python (abaixo), temos que $q=p=r=0.4226$

```python
from scipy.optimize import fsolve
import math

def equations(a):
    q, r, p = a
    return (-3*q - 3*r + 3*q*r + 2, 
            -3*p - 3*r + 3*p*r + 2,   
            -3*q - 3*p + 3*q*p + 2,
           )

q, r, p =  fsolve(equations, (0.4, 0.4, 0.4))

print(f"{q = :.4f}, {r = :.4f}, {p = :.4f}")
```

Mantendo a modelagem do Duopólio de Cournot para um caso de monopólio, temos que a produção total $Q = q_1$, ou seja, só uma firma produzirá, com custo $c_1 > 0$

O preço é dado por:
$$
p(q_1) = A - q_1
$$

O lucro desta firma será então

$$
\pi_1(q_1) = (A - q_1) * q_1 - c_1 * q_1
$$

o ponto ótimo do lucro será

$$
\begin{align*}
\frac{\delta\pi}{\delta q} &= A - 2q_1^* - c_1 = 0 \\
&\implies q_1^* = \frac{A - c_1}{2}
\end{align*}
$$

que é um ponto de máximo pois a derivada segunda é negativa.

O payoff, inserindo $q^*$ em $\pi_1(q_1)$ será:

$$
\pi_1(q^*) = Aq^* - {q^*}^2 - c_1q^* = {-q^*}^2 + q^*(A-c_1)
$$

como $(A-c_1) = 2q^*$, teremos:
$$
\pi_1(q^*) = -{q^*}^2 + q^*(2q^*) = {(q^*)}^2
$$

A quantidade produzida tenderá a zero.

Seja o preço dado por:
$$
p(Q_i) = A - Q = A - q_i - K_i
$$

onde K é a produção total descontado a firma $i$ ($K_i = \sum{q_j}; \forall j, j \neq i$). O somatório deste limite quando $n \to \infty$ é A, dado que Q é sempre menor que A (impedindo preços negativos). A imposição de preços estritamente positivos limita a produção total Q ao teto de A.

Dessa forma, para uma firma $i$:
o ponto ótimo do lucro será

$$
\begin{align*}
\frac{\delta\pi_i}{\delta q_i} &= -2q_i - c_i + (A-Q) = 0 \\
&\implies q_i^* = \frac{-c_i}{2} \to 0
\end{align*}
$$

sendo $q$ estritamente positivo, concluímos que conforme o número de participantes no mercado tende a infinito, a produção de cada um deles tenderá a zero conforme a produção agregada se aproxime da demanda A

```python
A =  130
c1 = c2 = 10

q1 = (A - 2*c1 + c2)/3
q2 = (A - 2*c2 + c1)/3
payoff_1 = (A - q1 - q2)*q1 - c1*q1
payoff_2 = (A - q2 - q1)*q2 - c2*q2
print(f"{q1 = }\n{q2 = }\n{payoff_1 = }\n{payoff_2 = }")

p1 = []
p2 = []
q1 = 50
q2 = 90
for i in range(0, 50):
    q1 = max(0, (A-c1-q2)/2)
    q2 = max(0, (A-c2-q1)/2)
    payoff_1 = (A - q1 - q2)*q1 - c1*q1
    payoff_2 = (A - q2 - q1)*q2 - c2*q2    
    p1.append(payoff_1)
    p2.append(payoff_2)


plt.plot(p1, label = '1')
plt.plot(p2, label = '2')
plt.legend()
plt.grid()
plt.show()
```

$$
