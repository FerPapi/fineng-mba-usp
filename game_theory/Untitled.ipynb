Para uma estratgia mista, se a probabilidades $(q, p, r)$ correspondem  probabilidade de cada jogador atuar, ento a estratgia mista pode se resumir ao sistema de equaes no lineares:
$$
-3q - 3r + 3qr + 2 = 0
$$

$$
-3p - 3r + 3pr + 2 = 0
$$

$$
-3q - 3p + 3qp + 2 = 0
$$

e resolvendo este sistema via python (abaixo), temos que $q=p=r=0.4226$

```python
from scipy.optimize import fsolve
import math

def equations(a):
    q, r, p = a
    return (-3*q - 3*r + 3*q*r + 2, 
            -3*p - 3*r + 3*p*r + 2,   
            -3*q - 3*p + 3*q*p + 2,
           )

q, r, p =  fsolve(equations, (0.4, 0.4, 0.4))

print(f"{q = :.4f}, {r = :.4f}, {p = :.4f}")
```

Mantendo a modelagem do Duoplio de Cournot para um caso de monoplio, temos que a produo total $Q = q_1$, ou seja, s uma firma produzir, com custo $c_1 > 0$

O preo  dado por:
$$
p(q_1) = A - q_1
$$

O lucro desta firma ser ento

$$
\pi_1(q_1) = (A - q_1) * q_1 - c_1 * q_1
$$

o ponto timo do lucro ser

$$
\begin{align*}
\frac{\delta\pi}{\delta q} &= A - 2q_1^* - c_1 = 0 \\
&\implies q_1^* = \frac{A - c_1}{2}
\end{align*}
$$

que  um ponto de mximo pois a derivada segunda  negativa.

O payoff, inserindo $q^*$ em $\pi_1(q_1)$ ser:

$$
\pi_1(q^*) = Aq^* - {q^*}^2 - c_1q^* = {-q^*}^2 + q^*(A-c_1)
$$

como $(A-c_1) = 2q^*$, teremos:
$$
\pi_1(q^*) = -{q^*}^2 + q^*(2q^*) = {(q^*)}^2
$$

Seja o preo dado por:
$$
p(Q_i) = A - Q = A - q_i - K_i

onde K  a produo total descontado a firma $i$ ($K_i = \sum{q_j}; \forall j, j \neq i$). O somatrio deste limite quando $n \to \infty$  A, dado que Q  sempre menor que A (impedindo preos negativos). A imposio de preos estritamente positivos limita a produo total Q ao teto de A.

Dessa forma, para uma firma $i$:
o ponto timo do lucro ser

$$
\begin{align*}
\frac{\delta\pi_i}{\delta q_i} &= -2q_i - c_i + (A-Q) = 0 \\

\end{align*}

sendo $q$ estritamente positivo, conclumos que conforme o nmero de participantes no mercado tende a infinito, a produo de cada um deles tender a zero conforme a produo agregada se aproxime da demanda A

$$
\begin{align*}
